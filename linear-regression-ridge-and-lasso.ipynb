```python
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_boston()
```

```python
type(df)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
```

```python
dataset.head()
```

```python
dataset['Price'] = df.target
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head(), y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
mse = cross_val_score(lin_reg,X,y,scoring = 'neg_mean_squared_error',cv = 10)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python
# Hyperparameter tuning

# alpha - learning rate
# max_iter - number of iterations
```

```python
params = {'alpha':[1e-5,1e-3,3e-3,1e-2,3e-2,1e-1,3e-1,1,3,10,30,100]}
ridge_regressor = GridSearchCV(ridge,params,scoring = 'neg_mean_squared_error',cv=10)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
```

```python
params = {'alpha':[1e-5,1e-3,3e-3,1e-2,3e-2,1e-1,3e-1,1,3,10,20]}
lasso_regressor = GridSearchCV(lasso,params,scoring = 'neg_mean_squared_error',cv=10)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

You can find out using all of them - Linear, Ridge, Lasso Regression

```python
y_pred = ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2 = r2_score(y_pred, y_test)
```
